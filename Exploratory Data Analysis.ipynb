```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
```

```python
df2000.head()
```

```python
df2001.head()
```

```python
df = pd.concat([df2000,df2001])
df.head()
```

```python
valid_budget_revenue = len(df[(df['budget']>0) & (df['revenue']>0)])
print(valid_budget_revenue)
```

```python
vbr_df = df[(df['budget']>0) & (df['revenue']>0)]
vbr_df.head()
```

```python
vbr_df['certification'].value_counts()
```

```python
vbr_df.groupby('certification', as_index=False)['revenue'].mean()
```

```python
vbr_df.groupby('certification', as_index=False)['budget'].mean()
```

```python
df.to_csv("Data/tmdb_results_combined.csv.gz",compression='gzip',index=False)
```
